3.  Is there a relationship between the percentage of dose 1 vaccination for staff and the percentage of dose 2 vaccination for students?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/vaccination/vax_school.csv')

# Create a scatter plot
plt.scatter(data['dose1_staff'], data['dose2_student'])
plt.xlabel('Percentage of Dose 1 Vaccination for Staff')
plt.ylabel('Percentage of Dose 2 Vaccination for Students')
plt.title('Relationship Between Dose 1 Vaccination for Staff and Dose 2 Vaccination for Students')

# Calculate the correlation coefficient
correlation = np.corrcoef(data['dose1_staff'], data['dose2_student'])[0, 1]
print(f"Correlation coefficient: {correlation}")

plt.show()

```

4. What is the correlation between doses given to staff and students?

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

correlation = data[['dose1_staff', 'dose2_staff', 'dose3_staff', 'dose1_student', 'dose2_student']].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Between Doses')
plt.show()
```

*   We can better grasp the relationship between the doses administered to staff and students with the aid of this visualisation. It provide insights into whether there is a regular pattern or connection between the various dosage forms that are used.
*   The degree and direction of the correlation between each pair of dose-related columns will be graphically represented in the resulting plot. Strong positive correlations are shown by values near 1, strong negative correlations by values near -1, and weak correlations by values near 0.

8. Is there any different between vaccination coverage between top 5 state and bottom 5 states?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'vax_cleaned'
# Compute vaccination coverage for each state
state_vaccination_coverage = vax_cleaned.groupby('state').apply(lambda x: ((x['dose1_staff'] + x['dose2_staff']).sum() / (x['dose1_student'] + x['dose2_student']).sum()) * 100)

# Sort states based on vaccination coverage in descending order
sorted_states = state_vaccination_coverage.sort_values(ascending=False)

# Get the top 5 and bottom 5 states
top_5_states = sorted_states.head(5)
bottom_5_states = sorted_states.tail(5)

# Plotting side by side for top 5 and bottom 5 states using Matplotlib
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Plotting the results for top 5 states
ax[0].bar(range(len(top_5_states)), top_5_states, align='center', alpha=0.7, color='green')
ax[0].set_xticks(range(len(top_5_states)))
ax[0].set_xticklabels(top_5_states.index, rotation='vertical')
ax[0].set_xlabel('State')
ax[0].set_ylabel('Vaccination Coverage (%)')
ax[0].set_title('Top 5 States')
for i, v in enumerate(top_5_states):
    ax[0].text(i, v + 1, f'{v:.2f}%', ha='center')

# Plotting the results for bottom 5 states
ax[1].bar(range(len(bottom_5_states)), bottom_5_states, align='center', alpha=0.7, color='red')
ax[1].set_xticks(range(len(bottom_5_states)))
ax[1].set_xticklabels(bottom_5_states.index, rotation='vertical')
ax[1].set_xlabel('State')
ax[1].set_ylabel('Vaccination Coverage (%)')
ax[1].set_title('Bottom 5 States')
for i, v in enumerate(bottom_5_states):
    ax[1].text(i, v + 1, f'{v:.2f}%', ha='center')

plt.tight_layout()
plt.show()
```

The result shows that state with high vaccination coverage is state 12
 with 110.87%. while lowest vaccination coverage is 15 with 104.01%. Thus, shows Malaysia provide good distribution of vaccine to public school.
